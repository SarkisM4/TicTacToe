```python
from IPython.display import clear_output

def Display_Board(Board):
    #clear_output()
    print("   |   |   ")
    print(' '+ Board[1] + ' | ' + Board[2] + ' | ' + Board[3])
    print("   |   |   ")
    print('------------')
    print("   |   |   ")
    print(' '+ Board[4] + ' | ' + Board[5] + ' | ' + Board[6])
    print("   |   |   ")
    print('------------')
    print("   |   |   ")
    print(' ' + Board[7] + ' | ' + Board[8] + ' | ' + Board[9])
    print("   |   |   ")


```

```python
def Player_Marker():
    
    marker = ' '
    
    while marker != 'X' and marker != 'O':
        
        marker = input("Player 1, choose to be X or O: ")
        
    Player1 = marker
        
    if Player1 == 'X':
        Player2 = 'O'
    
    else:
        Player2 = 'X'
        
    print("Player 1 is " + Player1)
    print("Player 2 is " + Player2)
    print('----------------')
    return (Player1, Player2)
```

```python
def Player_Choice():
    pos = 0
    
    while pos not in range(1,10):
        pos = int(input("Choose a position on the board: "))
        
    return pos

```

```python
def Game_On():
    choice = ' '
    
    while choice not in ['Y', 'y', 'N', 'n']:
        choice = input("Would you like to Continue Playing?(Y/N)")
        
    return choice
```

```python
def Win_Condition(Board, player):
    
    condition = False
    
    if Board[1] == Board[2] == Board[3] == player:
        condition = True
        
    elif Board[4] == Board[5] == Board[6] == player:
        condition = True
        
    elif Board[7] == Board[8] == Board[9] == player:
        condition = True
        
    elif Board[1] == Board[4] == Board[7] == player:
        condition = True
        
    elif Board[2] == Board[5] == Board[8] == player:
        condition = True
        
    elif Board[3] == Board[6] == Board[9] == player:
        condition = True
        
    elif Board[1] == Board[5] == Board[9] == player:
        condition = True
    
    elif Board[3] == Board[5] == Board[7] == player:
        condition = True
        
    return condition
```

```python
Board =  ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
choice = ' '

print("Welcome to TicTacToe!")
print('---------------------------------------')

#Display_Board(Board)
#player1, player2 = Player_Marker()

#print('---------------------------------------')

while choice != 'N' or choice != 'n':
    Board =  ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    choice = ' '
    
    Display_Board(Board)
    player1, player2 = Player_Marker()

    while ' ' in Board:
        print("Player 1's Turn")
        position = Player_Choice()
        Board[position] = player1
        Display_Board(Board)
    
        condition = Win_Condition(Board, player1)
        if condition == True:
            print("Player 1 Wins! ")
            break
        elif ' ' not in Board:
            print('Tie')
            break
        else:
            pass
            
      
        print("Player 2's Turn")
        position = Player_Choice()
        Board[position] = player2
        Display_Board(Board)
    
        condition = Win_Condition(Board, player2)
        if condition == True:
            print("Player 2 Wins!")
            break
        elif ' ' not in Board:
            print('Tie')
            break
        else:
            pass
    

    choice = Game_On()
    if choice == 'N' or choice == 'n':
        print('Thanks for Playing')
        break
    
    elif choice == 'Y' or choice == 'y':
        print('---------------------------------------')
        continue
```
